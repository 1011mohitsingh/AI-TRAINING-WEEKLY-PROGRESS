### Merging 2 dataframes

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

#DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees  
```

```python
salaries
```

```python
pd.merge(employees, salaries, on ='employee_id', how = 'inner') # here on is the basic of which column you are mergin and how is the inner means we had to take only the common one value
```

```python
pd.merge(employees,salaries,on='employee_id',how ='outer')  # it include either but in place of value nan comes
```

```python
pd.merge(employees,salaries,on='employee_id',how ='left')  # one the basis of the employee so we want all the value of the employee
```

```python
pd.merge(employees,salaries,on='employee_id',how ='right') # one the basis of the salary so we want all the value of the salary
```

##### generally we do the mergin with how as inner as we dont want anything means the value to be nan thats why

### Concatenation of 2 dataframes

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df1,df2], axis = 1)
```

### joining 2 data frames

```python
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2, how ='outer')
```

```python
df2.join(df1)
```
