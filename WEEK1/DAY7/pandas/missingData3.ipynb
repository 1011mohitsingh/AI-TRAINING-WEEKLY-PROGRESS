```python
import numpy as np 
import pandas as pd 
# fillna() for filling, for removing it is dropna() and for the searching is isna()
```

### find missing data

```python
data = {
    'A':[1,2,np.nan,4,5],  # np.nan is the null values
    'B':[1,2,3,4,5],
    'C':[1,2,3,np.nan,np.nan],
    'D':[1,np.nan,np.nan,np.nan,5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna() # It gives true for the null values and for rest false
```

```python
df.isna().sum()
```

```python
df.isna().any() # this says is there is any rows which has false value are all the rows means entire columns gives the true value
```

### Removing Missing Data

```python
df
```

##### when we have to delete the data we have to delete it using the rows okkay

```python
df.dropna()
```

```python
df
```

```python
df.dropna(thresh = 1) # here thresh is the no of non values present and that we are using for removing.
```

### find the missing data

```python
df.fillna(0)
```

```python
# fill and drop is on the basis of the column wise you know
values = {'A':0,'B':100,"C":300,'D':400}
df.fillna(value = values)
```
