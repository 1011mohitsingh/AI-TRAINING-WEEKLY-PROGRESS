```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df_index = pd.read_csv("economic_index.csv")
```

```python
df_index.head()
```

```python
# drop unnecessary columns
df_index.drop(columns=["Unnamed: 0", "year", "month"], inplace=True)
```

```python
# check null values
df_index.isnull().sum()
```

```python
## lets do some visualization to understand the data better
import seaborn as sns
sns.pairplot(df_index)
```

```python
df_index.corr()
```

```python
## visualize the datapoints more closely
plt.scatter(df_index['interest_rate'], df_index['unemployment_rate'], color ='r')
plt.xlabel("interest_rate")
plt.ylabel("unemployment_rate")
```

```python
### independent and dependent features
X = df_index.iloc[:,:-1]
y = df_index.iloc[:,-1]
```

```python
X.head()
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
import seaborn as sns
 
```

```python
sns.regplot(x=df_index['interest_rate'], y=df_index['index_price'])
```

```python
sns.regplot(x=df_index['interest_rate'], y=df_index['unemployment_rate'])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
### cross validation
from sklearn.model_selection import cross_val_score
validation_score = cross_val_score(regression, X_train, y_train, scoring='neg_mean_squared_error', cv =5)
```

```python
np.mean(validation_score)
```

```python
### prediction
y_pred = regression.predict(X_test)
```

```python
y_pred
```

```python
## performance Metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred)
print(score)
# display adjusted R-squared value
print(1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
```
