```python
import numpy as np
import pandas as pd 
```

```python
df = pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5,12,7]], columns=['cgpa', 'profile_score', 'lpa'])
```

```python
# Initialize weights and bias as w=0.1 and bias=0
# And provide the schema of our MLP as 1 I/P layer, 1 Hidden Layer and 1 Output Layer
# Therefore overall 9 parameters 
def initialize_parameters(layer_dims):
    np.random.seed(3)
    parameters = {}
    L = len(layer_dims)

    for l in range(1,L):
        parameters['W' + str(l)] = np.ones((layer_dims[l-1], layer_dims[l])) * 0.1
        parameters['b'+ str(l)] = np.zeros((layer_dims[l],1))

    return parameters
```

```python
initialize_parameters([2,2,1])
```

```python
def linear_forward(A_prev,W, b):
    Z = np.dot(W.T, A_prev) + b
    return Z
```

```python
## Forward propogation 
def L_layer_forward(X, parameters):
    A = X
    L = len(parameters)//2
    for l in range(1,L+1):
        A_prev = A
        W1 = parameters['W'+str(l)]
        b1 = parameters['b'+str(l)]
        # print("A"+str(l-1)+": ", A_prev)
        # print("W"+str(l)+": ", W1)
        # print("A"+str(l)+": ", b1)
        # print("--"*20)

        A = linear_forward(A_prev, W1,b1)
        # print("A"+str(l)+": ",A)
        # print("**"*20)
    return A, A_prev
```

```python
X = df[['cgpa', 'profile_score']].values[0].reshape(2,1) # Shape(no of features, no of training example)
y = df[['lpa']].values[0][0]

# paramter initialization
parameters = initialize_parameters([2,2,1])

y_hat,A1 = L_layer_forward(X, parameters)
```

```python
# Loss function = MSE
(y-0.32)**2
```

```python
y_hat = y_hat[0][0]
```

```python
A1
```

```python
error = y - y_hat[0][0]
```

```python
def update_parameters(parameters,y,y_hat, A1, X):
    parameters['W2'][0][0] = parameters['W2'][0][0] + (0.001 * 2 * error *A1[0][0])
    parameters['W2'][1][0] = parameters['W2'][1][0] + (0.001 * 2 * error *A1[1][0])
    parameters['b2'][0][0] = parameters['W2'][1][0] + (0.001 * 2 * error)

    parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * 2 * error *parameters['W2'][0][0]*X[0][0])
    parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * 2 * error *parameters['W2'][0][0]*X[1][0])
    parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * 2 * error *parameters['W2'][0][0])

    parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * 2 * error *parameters['W2'][1][0]*X[0][0])
    parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * 2 * error *parameters['W2'][1][0]*X[1][0])
    parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * 2 * error *parameters['W2'][1][0])
```

```python
update_parameters(parameters,y,y_hat,A1,X)
```

```python
## epochs implementation
parameters = initialize_parameters([2,2,1])
epochs = 5

for i in range(epochs):
    Loss = []
    for j in range(df.shape[0]):
        X = df[['cgpa','profile_score']].values[j].reshape(2,1) # shape(no of features, no of training example)
        y = df[['lpa']].values[j][0]

        # paramter initialization
        y_hat, A1 = L_layer_forward(X, parameters)
        y_hat = y_hat[0][0]

        update_parameters(parameters, y, y_hat, A1, X)
        Loss.append((y-y_hat)**2)

    print('Epoch - ', i+1,'Loss - ', np.array(Loss).mean())
parameters
```
